```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import LabelEncoder

# Load data from the CSV file
file_path = '/mnt/d/20232024-01-CT252-B2003731_final/deep-metagnomics-pub-main-final/results/Quin_gut_liver_cirrhosis_binGMM_nb10_aun_0.0_1.0/details/a1_estopc5_rf_500_-2_o2adam_lr-1.0de0.0e500_20231127_091613c1.0di-1ch3details_s1k1_importance_fea.csv'
data = pd.read_csv(file_path, header=None)  # Assuming no header row in your data

# Extracting data from the first and second columns
names = data.iloc[:, 0]  # First column
fractions = data.iloc[:, 1]  # Second column

# Create a DataFrame from the extracted data
df = pd.DataFrame({'Species': names, 'Fraction': fractions})

# Khi to LabelEncoder
label_encoder = LabelEncoder()

# Chuyn i ct 'Species' thnh cc s nguyn
df['Species Index Encoded'] = label_encoder.fit_transform(df['Species'])

num_clusters = 4 

# S dng GMM
gmm = GaussianMixture(n_components=num_clusters, random_state=42)
gmm.fit(df[['Species Index Encoded', 'Fraction']])
labels = gmm.predict(df[['Species Index Encoded', 'Fraction']])

# V biu  scatter vi cm mu
plt.figure(figsize=(10, 6))
plt.scatter(df['Species Index Encoded'], df['Fraction'], c=labels, cmap='viridis', s=50)
plt.xlabel('Species Index Encoded')
plt.ylabel('Fraction')
plt.title('Cluster bacterial species in metahitIBD data using GMM.')
plt.colorbar(label='Cluster')
plt.show()
```
